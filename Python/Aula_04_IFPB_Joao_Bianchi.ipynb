```python
# continue e break
while True:
  tempo = int(input("Quantos anos de estudo você tem?"))
  if tempo > 0:
     break

for _ in range(tempo):
  print("Cachorro")
```

```python
def main():
  imprimir_cachorro(obter_numero())

def obter_numero():
  while True:
    n= int(input("Quantas vezes vc quer imprimir a palavra cachorro? "))
    if n> 1:
      return n

def imprimir_cachorro(a):
  for _ in range(a):
    print("Cachorro")

main()
```
