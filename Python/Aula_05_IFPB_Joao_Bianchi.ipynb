```python
#Usando listas
students = ["João", "Maria", "José", "Ana"]
print(students[0])
print(students[1])
print(students[2])
```

```python
for student in students:
  print(student)
```

```python
# adicionar no final da lista

students.append("Carlos")
print(len(students))

# O código abaixo faz a mesma coisa que students.append

students[len(students):] = ["Madalena"]

print(len(students))
```

```python
#remover item da lista
students.remove("Carlos")
```

```python
#contagem
students.count("João")
```

```python
#inverter ordem da lista
students.reverse()
```

```python
#Limpa a lista
students.clear()
```

```python
# Usando dicionários
students = ["João", "Maria", "José", "Ana", "Madalena"]
classes = ["Math", "English", "Portuguese", "Arts", "Science"]

students_class = {
    "João" : "Math",
    "Maria" : "English",
    "José" : "Portuguese",
    "Ana" : "Arts",
    "Madalena" : "Science"
}

for student in students_class:
  print(f"{student} : {students_class[student]}")
```

```python
joao_contact = {
    "id": 1,
    "nome": "João",
    "Telefone": "90000-0000"
}

maria_contact = {
    "id": 2,
    "nome": "Maria",
    "Telefone": "91111-1111"
}

people_list = [joao_contact, maria_contact]

for person in people_list:
  print(person)
```

```python
def main():
  print_column(10)

def print_column(height):
  for _ in range(height):
    print("#")

main()
```

```python
def main():
  print_row(10)
  print_blank(50)

def print_row(widht):
    print("-" * widht)

def print_blank(blank):
  print(" " * blank)

main()
```

```python
def main():
  print_square(10)

def print_square(size):
  #para cada linha do quadrado
  for i in range(size):
    #para cada coluna do quadrado
    for j in range(size):
      print("#", end="")
    print()
main()
```
