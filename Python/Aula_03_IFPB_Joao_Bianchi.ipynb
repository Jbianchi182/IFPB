```python
name = input("What's your city's name? ").strip().title()

if name == "João Pessoa" or name == "Campina Grande" or name == "Patos":
  print("You live in PB")
elif name == "Recife":
  print("You live in PE")
else:
  print("State not found")
```

```python
name = input("What's your city's name? ").strip().title()

match name:
  case "João Pessoa" | "Campina Grande" | "Patos":
    print("You live in PB")
  case "Recife":
    print("You live in PE")
  case _:
    print("State not found")
```
