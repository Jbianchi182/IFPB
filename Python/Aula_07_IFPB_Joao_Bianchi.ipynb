```python
#Libraries - Random

import random

print(random.choice(["Cara", "Coroa"]))
```

```python
coin = random.choice(["Cara", "Coroa"])
print(coin)
```

```python
from random import choice

coin = choice(["Cara", "Coroa"])
print(coin)
```

```python
number = random.randint(1,10)
print(number)
```

```python
cards = ["Valete", "Rainha", "Rei"]
random.shuffle(cards)

for card in cards:
  print(card)
```

```python
import json

termo = "weezer"
response = requests.get("https://itunes.apple.com/search?entity=song&limit=1&term="+termo)

print(json.dumps(response.json(), indent=2))
```

```python
import requests

termo = "weezer"
response = requests.get("https://itunes.apple.com/search?entity=song&limit=10&term="+termo)

output = response.json()

for resultado in output["results"]:
  print(resultado["trackName"])
```
